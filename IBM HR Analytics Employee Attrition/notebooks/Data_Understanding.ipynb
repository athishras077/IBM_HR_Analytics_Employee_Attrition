```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"D:\Data Analysis Project\IBM HR Analytics Employee Attrition\Data\raw\IBM HR Analytics Employee Attrition.csv")
```

```python
# Display the First Few Records and Last Few Records

print(df.head(5).to_string())
print()
print(df.tail(5).to_string())
```

```python
# Dataset Dimensions 

print(df.shape)
```

```python
# View Column Names

print(df.columns.tolist())
```

```python
# Dataset Information

print(df.info())
```

```python
# Summary Statistics

print(df.describe().to_string())
print()

# To summarize categorical columns

print(df.describe(include="object").to_string())
```

```python
# Before analysis, categorize each feature based on its data type and business meaning
# This helps determine appropriate analysis techniques and visualizations

# Numerical Features

numerical_columns = df.select_dtypes(include=["int64","float64"]).columns
print("Numbercal Columns: ")
print(numerical_columns.tolist())
```

```python
# Categorical Features

Categorical_columns = df.select_dtypes(include="object").columns
print("Categorical columns :")
print(Categorical_columns.tolist())
```

```python
# Check Missing Values

missing_values = df.isnull().sum()

missing_values_per = (missing_values/len(df)*100).round(2)

summary = (
    pd.DataFrame({
        "Missing Values": missing_values,
        "Missing Values Percentage": missing_values_per
    })
    .sort_values(by="Missing Values", ascending=False)
)

summary[summary["Missing Values"] > 0]
```

```python
# Check Duplicate Records

duplicate_rows = df.duplicated().sum()

print(f"Duplicate Rows: {duplicate_rows}")
```

```python
# Verify Employee IDs Are Unique

duplicate_employee_id = df['EmployeeNumber'].duplicated().sum()
print(f"Duplicate Employee IDs: {duplicate_employee_id}")
```

```python
# Unique Values in Each Column

unique_value = pd.DataFrame({
    "Coulmn": df.columns,
    "Unique Values":  df.nunique()
})
unique_value.sort_values("Unique Values")
```

```python
# Inspect Categorical Values
# Display Unique Values for Categorical Columns

categorical_columns = df.select_dtypes(include="object").columns

for column in categorical_columns:
    print("=" * 50)
    print(f"{column}")
    print(df[column].unique())
    print()

```

```python
# Frequency Distribution of Categories
# Understanding how categories are distributed helps identify class imbalance and unexpected values.

for column in categorical_columns:
    print("=" * 60)
    print(f"{column}")
    print(df[column].value_counts())
    print()
```

```python
# Validate Numeric Ranges
# # Check for Negative Values in Numeric Columns

numerical_columns = df.select_dtypes(include=["int64", "float64"]).columns

for column in numerical_columns:
    negative_count = (df[column] < 0).sum()

    if negative_count > 0 :
        print(f"{column}: {negative_count} Negative Values")
```

```python
# Create a Data Quality Summary Table

Data_Quality_Summary = pd.DataFrame({
    "Metric" : [
        "Total Rows",
        "Total Columns",
        "Missing Values",
        "Duplicate Rows",
        "Duplicate Employee IDs",
        "Constant Columns" 
    ],
    "Results" : [
        len(df),
        len(df.columns),
        int(df.isnull().sum().sum()),
        int(df.duplicated().sum()),
        int(df['EmployeeNumber'].duplicated().sum()),
        (df.nunique() == 1).sum()
    ]
})

print(Data_Quality_Summary)
```

```python
# ==========================================================
# Create a Working Copy of the Dataset
# ==========================================================

# Create a deep copy of the original DataFrame

df_clean = df.copy()

print(f"Original Dataset : {df.shape}")
print(f"Copied Dataset : {df_clean.shape}")
```

```python
# ==========================================================
# Data Cleaning & Preparation
# ==========================================================


# Identify Constant Columns

Constant_Column = df_clean.columns[df_clean.nunique() == 1].tolist()
print("Constant Columns : ")
print(Constant_Column)
print()
```

```python
# Remove Constant Columns

df_clean.drop(columns= Constant_Column, inplace=True)
print(df_clean.shape)
```

```python
remaining_constant_columns = df_clean.columns[df_clean.nunique() == 1].tolist()

print("Remaining Constant Columns:")
print(remaining_constant_columns)
```

```python
# Check whether EmployeeNumber exists
print('EmployeeNumber' in df_clean.columns)
```

```python
# Remove EmployeeNumber

df_clean.drop(columns='EmployeeNumber', inplace=True)
print("Updated Shape:", df_clean.shape)
```

```python
# Verify and Optimize Data Types
# Now that we've cleaned unnecessary columns, we need to ensure each column has the correct data type.

print(df_clean.info())
```

```python
# Count Each Data Type

print(df_clean.dtypes.value_counts())
df_clean.select_dtypes(include="object").columns.tolist()
```

```python
# Converting object datatype into category datatype

object_columns = df_clean.select_dtypes(include="object").columns
df_clean[object_columns] = df_clean[object_columns].astype("category")
```

```python
# Verify the result

print(df_clean.dtypes.value_counts())
len(df_clean.select_dtypes(include="category").columns)
```

```python
# Final Dataset Validation

print("Shape :", df_clean.shape)

print("\nMissing Values:")
print(df_clean.isnull().sum().sum())

print("\nDuplicate Rows:")
print(df_clean.duplicated().sum())

print("\nRemaining Constant Columns:")
print(df_clean.columns[df_clean.nunique()==1].tolist())
```

```python
# ==========================================================
# Save the Cleaned Dataset
# ==========================================================


df_clean.to_csv("IBM_HR_Analytics_Cleaned.csv", index=False)
print("Cleaned dataset saved successfully!")
```
